```python
df.drop('pool' , axis=1 , inplace = True)
df.drop('basement' , axis=1 , inplace = True)
df.drop('fireplace' , axis=1 , inplace = True)
df.drop('elevator' , axis=1 , inplace = True)
df.drop('stairs' , axis=1 , inplace = True)
```

```python
city_count = df['city'].value_counts()
city_count
```

```python
dis = df['district'].value_counts().reset_index()
dis.columns = ['district' , 'count']
dis
```

```python
df['size'].nunique()
```

```python
def get_price(price):
      if price <= 1100:
        return 'ممتازه'
      elif 2000 < price <= 10000:
        return 'جيده جيدا'
      elif 10000 < price <= 30000:
        return 'جيدا'
      elif 30000 < price <= 50000:
        return 'مقبوله'
      elif 50000 < price <= 100000:
        return 'مقبوله لحد ما'
      elif 100000 < price <= 200000:
        return 'سيئه'
      else:
        return 'مبالغ في سعرها'
df['recomanded_price'] = df['price'].apply(get_price)
```

```python
high_dis = df.groupby('district')[['size' , 'price']].sum().reset_index().sort_values('size' , ascending=False)
high_dis
```
